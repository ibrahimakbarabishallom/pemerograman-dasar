```python
# SOAL 2: Menghitung Volume Kubus
def hitung_volume_balok(panjang, lebar, tinggi):
    """Menghitung volume balok dari panjang, lebar, dan tinggi."""
    return panjang * lebar * tinggi

# Pemanggilan:
print(f"Volume balok adalah: {hitung_volume_balok(6, 4, 2)}")

```

```python
# SOAL 3: Menghitung Rata-rata dari List Nilai
def hitung_rata_rata(data_nilai):
    """Menghitung rata-rata dari list angka dengan validasi nilai."""
    total = 0
    jumlah_valid = 0

    for nilai in data_nilai:
        if 0 <= nilai <= 100:     # hanya nilai valid yang dihitung
            total += nilai
            jumlah_valid += 1
        else:
            print(f"Nilai {nilai} diabaikan (tidak valid).")

    if jumlah_valid > 0:
        return total / jumlah_valid
    return 0

# Contoh pemanggilan:
nilai_praktikan = [85, 110, -5, 90, 78]
print(f"Nilai rata-rata kelas adalah: {hitung_rata_rata(nilai_praktikan)}")
```

```python
# SOAL 4: Sistem Penentuan Predikat Nilai
def tentukan_predikat(skor):
    """Menentukan predikat (A, B, C, D, E) berdasarkan skor."""

    # Pengecekan tipe data agar hanya menerima angka
    if not isinstance(skor, (int, float)):
        return "Input harus berupa angka!"

    # Logika penentuan predikat
    if skor >= 85:
        return "A"
    elif skor >= 70:
        return "B"
    elif skor >= 60:
        return "C"
    elif skor >= 50:
        return "D"
    else:
        return "E"


# Pemanggilan contoh
print(f"Skor 75 mendapat predikat: {tentukan_predikat(75)}")
print(f"Skor 48 mendapat predikat: {tentukan_predikat(48)}")
print(f"Skor 'abc' mendapat predikat: {tentukan_predikat('abc')}")
```

```python
# SOAL 5: Menghitung Faktorial dengan Loop While
def hitung_faktorial(n):
    """Menghitung faktorial dari bilangan positif n."""

    # Tangani kasus khusus
    if not isinstance(n, int):
        return "Input harus berupa bilangan bulat!"
    if n < 0:
        return "Faktorial tidak didefinisikan untuk bilangan negatif."
    if n == 0:
        return 1

    hasil = 1

    # Perulangan berjalan selama n lebih besar dari 1
    while n > 1:
        hasil *= n
        n -= 1  # Kurangi n setiap iterasi

    return hasil


# Pemanggilan:
bilangan_n = 4
print(f"Faktorial dari {bilangan_n} adalah: {hitung_faktorial(bilangan_n)}")  # Output: 24

# Coba kasus tambahan
print(hitung_faktorial(0))   # Output: 1
print(hitung_faktorial(-3))  # Output: Faktorial tidak didefinisikan untuk bilangan negatif.
print(hitung_faktorial(5))   # Output: 120
```

```python
# SOAL 6: Analisis Data Suhu Ekstrem (for dan if/elif)
def analisis_suhu(data_suhu, batas_panas=35, batas_dingin=15):
    """Menghitung jumlah hari Panas dan Dingin dari data_suhu."""

    # Inisialisasi penghitung
    panas_count = 0
    dingin_count = 0

    # Perulangan untuk setiap data suhu
    for suhu in data_suhu:
        if suhu >= batas_panas:
            panas_count += 1
        elif suhu <= batas_dingin:
            dingin_count += 1

    # Mengembalikan hasil
    return panas_count, dingin_count


# Pemanggilan:
suhu_mingguan = [36, 30, 14, 25, 38, 16, 28]

jumlah_panas, jumlah_dingin = analisis_suhu(suhu_mingguan)

print(f"Jumlah hari Panas (≥{35}°C): {jumlah_panas}")
print(f"Jumlah hari Dingin (≤{15}°C): {jumlah_dingin}")
```
