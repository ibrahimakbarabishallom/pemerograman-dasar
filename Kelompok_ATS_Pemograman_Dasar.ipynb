```python
kapasitas = 0   # mulai dari 0 liter

while kapasitas <= 100:
    print("Isi sekarang:", kapasitas, "liter")

    if kapasitas == 80:
        print("WARNING: Hampir penuh!")
    if kapasitas == 100:
        print("TANGKI PENUH – Pengisian berhenti.")
        break  # hentikan pengisian

    kapasitas += 10   # tambah 10 liter setiap iterasi
```

Penjelasan Singkat :
Program mulai dari 0 liter.
Selama belum mencapai 100 liter, program terus mengisi dengan menambah 10 liter tiap putaran while.
Ketika volume mencapai 80 liter, program memberi pesan “WARNING: Hampir penuh!”.
Saat volume mencapai 100 liter, program menampilkan “TANGKI PENUH – Pengisian berhenti.” dan pengisian selesai.
