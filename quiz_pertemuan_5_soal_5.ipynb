```python
n = 5
faktorial = 1
i = n

while i > 0:
    faktorial *= i
    i -= 1

print("Faktorial dari", n, "adalah:", faktorial)
```

```python
total = 0
for i in range(1, 101):
    if i % 2 == 0:
        total += i
print("Jumlah semua bilangan genap dari 1 sampai 100 adalah:", total)
```
