```python
angka = [2, 4, -1, 6, 8]

for nilai in angka:
    if nilai < 0:
        print("Angka negatif ditemukan!")
        break
    print(f"Angka positif: {nilai}")
```

```python
for i in range(10):
    if i % 2 == 0:  # Lewati angka genap
        continue
    print(f"Angka ganjil: {i}")
```

```python
for nilai in range(10):
    print("Hello Polibatam!")
    if nilai == 5:
        break
else:
    print("Iterasi FOR selesai hingga akhir")
```
