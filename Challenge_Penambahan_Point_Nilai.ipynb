```python
# ============================
#   FUNGSI MENGHITUNG ENERGI
# ============================
def hitung_energi(daya_watt, jam):
    # Energi (kWh) = (Watt Ã— Jam) / 1000
    return (daya_watt * jam) / 1000


# ============================
#       PROGRAM UTAMA
# ============================
print("=== Perhitungan Konsumsi Energi Listrik ===")

total_peralatan = int(input("Berapa banyak peralatan listrik? "))

# Validasi input minimal 1 (while)
while total_peralatan <= 0:
    print("Jumlah peralatan tidak boleh 0 atau negatif!")
    total_peralatan = int(input("Masukkan ulang jumlah peralatan: "))

total_energi = 0  # accumulator

# Perulangan for
for i in range(1, total_peralatan + 1):
    print(f"\nPeralatan ke-{i}")
    daya = float(input("  Masukkan daya (Watt): "))
    waktu = float(input("  Lama pemakaian (jam): "))

    energi = hitung_energi(daya, waktu)
    print(f"  Energi peralatan ini = {energi:.2f} kWh")

    total_energi += energi  # akumulasi


# ============================
#     PERCABANGAN LOGIKA
# ============================
print("\n=== HASIL TOTAL ===")
print(f"Total energi harian = {total_energi:.2f} kWh")

if total_energi < 1:
    print("Kategori: Rendah (hemat energi)")
elif total_energi <= 5:
    print("Kategori: Sedang")
else:
    print("Kategori: Tinggi (boros energi!)")
```
